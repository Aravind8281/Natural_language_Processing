```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Sample sentences for training
sentences = [
    'I love cats theyre very cute and kind'
    'Word embeddings are powerful.',
    'Neural networks can learn embeddings.',
    'Embeddings capture semantic relationships.',
]

# Labels corresponding to each sentence (assuming you have 4 classes)
labels = np.array([0, 1, 2, 3])

# Tokenize and pad the sequences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
total_words = len(tokenizer.word_index) + 1

input_sequences = tokenizer.texts_to_sequences(sentences)
max_sequence_length = max(len(seq) for seq in input_sequences)

padded_sequences = pad_sequences(input_sequences, maxlen=max_sequence_length, padding='pre')

# Prepare input and output for the model
X = padded_sequences[:, :-1]
y = labels  # Assuming you have one label for each sentence

# Build the model
model = Sequential()
model.add(Embedding(input_dim=total_words, output_dim=10, input_length=max_sequence_length - 1))
model.add(LSTM(100, return_sequences=True))
model.add(LSTM(100, return_sequences=True))
model.add(LSTM(100))
model.add(Dense(4, activation='softmax'))  # Output dimension adjusted to 4

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Train the model
model.fit(X, y, epochs=100, verbose=1)

# Save the trained model
model.save('word_embedding_model.h5')
```

```python
from tensorflow.keras.models import load_model
import numpy as np

# Load the trained model
model = load_model('word_embedding_model.h5')

# Tokenize and pad the input query
input_query = 'How do word embeddings work?'
input_seq = tokenizer.texts_to_sequences([input_query])
padded_seq = pad_sequences(input_seq, maxlen=X.shape[1])

# Predict the next words
predicted_probabilities = model.predict(padded_seq, verbose=0)

# Find the index of the word with the highest probability
predicted_indices = np.argmax(predicted_probabilities, axis=-1)

# Convert predicted indices back to words
predicted_words = [tokenizer.index_word[idx] for idx in predicted_indices.flatten()]

# Generate the response sentence
response_sentence = ' '.join(predicted_words)

print(f'Input Query: {input_query}')
print(f'Response Sentence: {response_sentence}')
```
